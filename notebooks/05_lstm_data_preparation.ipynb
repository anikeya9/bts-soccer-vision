```python
import pandas as pd
import numpy as np
import glob
import re
```

```python
path = "lstm_data/video*.parquet"

file_list = glob.glob(path)
```

```python
file_list
```

```python
data_frames = {}
for file in file_list:
    # Extract the key from the filename
    key1 = re.split(r'[_ .]',file)[2]
    key2 = re.split(r'[_ .]',file)[4]
    data = pd.read_parquet(file)
    data_frames[f"{key1}_{key2}"] = data
```

```python
data_frames.keys()
```

```python
s = 0
for key in data_frames.keys():
    le = len(data_frames[key])
    s = s+le
    
```

```python

# Concatenate all DataFrames
df = pd.concat(data_frames.values(), ignore_index=True)


```

```python
df
```

```python
# Find the number of rows without any NaN values 
rows_without_nan = df.dropna().shape[0] 
print(f"Number of rows without any NaN values: {rows_without_nan}")
```

```python
new_df = df.dropna()
```

```python
new_df.to_parquet("lstm_data/lstm_data.parquet")
```
